```python
import xml.etree.ElementTree as ET
import glob
import os
import json
import shutil
```

```python
def xml_to_yolo_bbox(bbox, w, h):
    # xmin, ymin, xmax, ymax
    x_center = ((bbox[2] + bbox[0]) / 2) / w
    y_center = ((bbox[3] + bbox[1]) / 2) / h
    width = (bbox[2] - bbox[0]) / w
    height = (bbox[3] - bbox[1]) / h
    return [x_center, y_center, width, height]
```

```python
train_image_dir = "Tensorflow/workspace/training/images/train"
test_image_dir = "Tensorflow/workspace/training/images/test"

os.mkdir('yolov4/images')
```

```python
files = glob.glob(os.path.join(train_image_dir, '*.xml'))

for file in files:
    tree=ET.parse(file)
    root=tree.getroot()
    
    width=int(root.find('size').find('width').text)
    height=int(root.find('size').find('height').text)
    filename=root.find('filename').text.split('.')[0]
    classes=[]
    result=[]
    
    shutil.copy(os.path.join('Tensorflow/workspace/training/images/train',filename+'.jpg'),'yolov4/images')
    for obj in root.findall('object'):
        

        classname=obj.find('name').text
        if classname not in classes:
            classes.append(classname)
        xmin=int(obj.find('bndbox').find('xmin').text)
        ymin=int(obj.find('bndbox').find('ymin').text)
        xmax=int(obj.find('bndbox').find('xmax').text)
        ymax=int(obj.find('bndbox').find('ymax').text)

        bbox=[xmin,ymin,xmax,ymax]

        yolo_bbox=xml_to_yolo_bbox(bbox,width,height)

        index=classes.index(classname)
        bbox_string="".join([str(x) for x in yolo_bbox])
        result.append(f"{index} {bbox_string}")
    if result:
        with open(os.path.join('yolov4/images',filename+".txt"),'w') as f:
            f.write("\n".join(result))
    
    
    
```
